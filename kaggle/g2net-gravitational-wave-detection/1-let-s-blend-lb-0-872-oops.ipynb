```python
sub1 = pd.read_csv('../input/g2net-eda-and-modeling/model_submission.csv').sort_values('id') 
sub2 = pd.read_csv('../input/k/yasufuminakama/g2net-efficientnet-b0-baseline-inference/submission.csv').sort_values('id') 
sub3 = pd.read_csv('../input/tf-g2net-eda-and-starter/submission.csv').sort_values('id') 
sub4 = pd.read_csv('../input/g2net-efficientnetb1-tpu-evaluate/submission.csv').sort_values('id') 
sub5 = pd.read_csv('../input/g2net-fast-and-almost-accurate-detector/submission.csv').sort_values('id') 
sub6 = pd.read_csv('../input/cqt-g2net-efficientnetb1-tpu-inference/submission.csv').sort_values('id') 
```

```python
# Group data together
hist_data = [sub1.target, sub2.target, sub3.target, sub4.target, sub5.target, sub6.target]

group_labels = ['Yaroslav Isaienkov - 0.864', 'Y.Nakama - 0.860', 'Mrigendra Agrawal - 0.855', 'Welf Crozzo - 0.834',
                'kkiller - 0.855', 'Welf Crozzo CQT - 0.865']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2, show_hist=False, show_rug=False)
fig.show()
```

```python
data=np.corrcoef([sub1.target, sub2.target, sub3.target, sub4.target, sub5.target, sub6.target])
fig = px.imshow(data,
                x=group_labels,
                y=group_labels
               )
fig.show()
```
