```python
with open("../input/hashcode-2021-oqr-extension/hashcode.in", "r") as file:
    first = file.readline().split()
    duration, intersections_num, streets_num, cars_num, default_score = [int(x) for x in first]
    
    streets = {}
    for _ in range(streets_num):
        line = file.readline().split()
        start = int(line[0])
        end = int(line[1])
        street_name = line[2]
        length = int(line[3])
        streets[street_name] = start,end,length
        
    cars = []
    for _ in range(cars_num):
        line = file.readline().split()
        sequence = line[1:]
        cars.append(sequence)
```

```python
impossible_cars = 0  # number of cars that could not finish
total_required_distance = 0  # total distance to be travelled by the car

for car in cars:
    distance = 0
    for street_name in car[1:]:  # [1:] because the car does not need to travel along the first street
        distance += streets[street_name][2]

    if distance > duration:
        impossible_cars += 1
    else:
        total_required_distance += distance
        
impossible_cars
```

```python
unused_streets = streets.copy()

for car in cars:
    for street in car:
        unused_streets.pop(street, None)

used_streets = {k: v for k, v in streets.items() if k not in unused_streets}
```
